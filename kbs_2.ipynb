```python
from experta import Fact, Rule, KnowledgeEngine
```

```python
user_answers = {}
```

```python
class StudentFacts(Fact): pass
```

```python
class CareerRecommender(KnowledgeEngine):
    def __init__(self):
        super().__init__()
        self.recommendation = None

    @Rule(StudentFacts(personality='logical'),
          StudentFacts(subject='math'),
          StudentFacts(gpa='yes'),
          StudentFacts(interest='tech'),
          StudentFacts(skill='problem-solving'),
          StudentFacts(motivation='inventing'),
          StudentFacts(goal='stable'))
    def recommend_engineering(self):
        self.recommendation = "Engineering"

    @Rule(StudentFacts(personality='emotional'),
          StudentFacts(subject='biology'),
          StudentFacts(gpa='yes'),
          StudentFacts(interest='people'),
          StudentFacts(skill='communication'),
          StudentFacts(motivation='helping'),
          StudentFacts(goal='help others'))
    def recommend_medical(self):
        self.recommendation = "Medical"

    @Rule(StudentFacts(subject='accounting'),
          StudentFacts(skill='math'),
          StudentFacts(goal='stable'))
    def recommend_accounting(self):
        self.recommendation = "Accounting"

    @Rule(StudentFacts(motivation='exploring'),
          StudentFacts(environment='outdoor'))
    def recommend_explorer(self):
        self.recommendation = "Explorer (Field Research or Nature Work)"

# Questions Logic
questions = [
    {
        "key": "personality",
        "question": "What's your personality type?",
        "options": ['logical', 'emotional', 'social', 'practical']
    },
    {
        "key": "subject",
        "question": "What's your favorite subject?",
        "branch_on": "personality",
        "branches": {
            "logical": ['math', 'physics', 'computer'],
            "emotional": ['biology', 'literature'],
            "default": ['math', 'biology', 'history', 'accounting']
        }
    },
    {
        "key": "gpa",
        "question": "Is your GPA high (above 3.5)?",
        "options": ['yes', 'no', 'maybe']
    },
    {
        "key": "interest",
        "question": "Do you prefer working with people or technology?",
        "options": ['people', 'tech', 'both']
    },
    {
        "key": "skill",
        "question": "Whatâ€™s your strength?",
        "options": ['problem-solving', 'creativity', 'communication', 'math']
    },
    {
        "key": "motivation",
        "question": "What motivates you more?",
        "options": ['helping', 'inventing', 'exploring']
    },
    {
        "key": "goal",
        "question": "Whatâ€™s your long-term goal?",
        "options": ['stable', 'help others', 'research', 'own business']
    },
    {
        "key": "environment",
        "question": "Where do you prefer to work?",
        "options": ['indoor', 'outdoor', 'lab']
    }
]

current_q = 0

def ask_question(index=0):
    global current_q
    current_q = index
    clear_output()

    if index >= len(questions):
        run_engine()
        return

    q = questions[index]
    key = q["key"]
    question = q["question"]

    # Branching options
    if "branch_on" in q:
        branch_key = q["branch_on"]
        branch_val = user_answers.get(branch_key, "default")
        options = q["branches"].get(branch_val, q["branches"]["default"])
    else:
        options = q["options"]

    # Create widgets
    label = widgets.Label(value=question)
    dropdown = widgets.Dropdown(options=options, description='Answer:')
    button = widgets.Button(description="Next", button_style='success')

    def on_click(b):
        user_answers[key] = dropdown.value
        ask_question(index + 1)

    button.on_click(on_click)

    display(label, dropdown, button)

def run_engine():
    clear_output()
    engine = CareerRecommender()
    engine.reset()
    engine.declare(StudentFacts(**user_answers))
    engine.run()

    rec = engine.recommendation or "No clear recommendation. Try different answers."
    display(widgets.HTML(f"<h3>Your Recommended Career Path: <span style='color:blue'>{rec}</span></h3>"))

# Start
ask_question(0)
```

```python
# Install necessary libraries
!pip install ipywidgets --quiet
from IPython.display import display, clear_output
import ipywidgets as widgets

# Store answers here
answers = {}

# Define branching questions
questions = [
    {
        "text": "What best describes your personality?",
        "options": ["Logical", "Emotional", "Social", "Practical"],
        "key": "personality"
    },
    {
        "text": "What is your favorite subject?",
        "options": {
            "Logical": ["Math", "Physics", "Computer Science"],
            "Emotional": ["Biology", "Chemistry", "Psychology"],
            "Social": ["Economics", "History", "Literature"],
            "Practical": ["Agriculture", "Business Studies", "Accounting"]
        },
        "key": "subject"
    },
    {
        "text": "Do you like working with people or technology?",
        "options": {
            "Logical": ["Technology", "Both"],
            "Emotional": ["People", "Both"],
            "Social": ["People", "Both"],
            "Practical": ["Hands-on", "People"]
        },
        "key": "interest"
    },
    {
        "text": "Is your GPA above 3.5?",
        "options": ["Yes", "No"],
        "key": "gpa"
    },
    {
        "text": "What environment do you prefer to work in?",
        "options": ["Indoor", "Outdoor", "Lab"],
        "key": "environment"
    },
    {
        "text": "What motivates you more?",
        "options": ["Helping others", "Inventing", "Exploring", "Earning money"],
        "key": "motivation"
    }
]

# Recommendation logic
def get_recommendation(ans):
    if ans['personality'] == 'Logical' and ans['subject'] in ['Math', 'Physics', 'Computer Science'] and ans['interest'] in ['Technology', 'Both']:
        return "Engineering or Computer Science"
    elif ans['personality'] == 'Emotional' and ans['subject'] in ['Biology', 'Chemistry', 'Psychology']:
        return "Medical or Psychology"
    elif ans['personality'] == 'Social' and ans['interest'] == 'People':
        return "Business, Law, or Education"
    elif ans['personality'] == 'Practical' and ans['subject'] == 'Agriculture':
        return "Agricultural Sciences"
    elif ans['subject'] == 'Accounting':
        return "Accounting or Finance"
    else:
        return "Explorer or Interdisciplinary Fields (e.g., Research, Journalism)"

# UI Flow
question_index = 0

def ask_question(index):
    clear_output()
    q = questions[index]

    # Handle branching options
    if isinstance(q['options'], dict):
        personality = answers.get("personality", "")
        options = q['options'].get(personality, [])
    else:
        options = q['options']

    radio = widgets.RadioButtons(options=options, description='', style={'description_width': 'initial'})
    btn = widgets.Button(description="Next", button_style='info')

    def on_click(b):
        answers[q['key']] = radio.value
        next_index = index + 1
        if next_index < len(questions):
            ask_question(next_index)
        else:
            show_result()

    btn.on_click(on_click)
    display(widgets.Label(value=f"Q{index+1}: {q['text']}"))
    display(radio)
    display(btn)

def show_result():
    clear_output()
    rec = get_recommendation(answers)
    display(widgets.HTML(f"<h3>Your Recommended Career Path:</h3><p><b>{rec}</b></p>"))
    display(widgets.HTML("<h4>Thank you for using the system!</h4>"))

# Start the questionnaire
ask_question(0)
```
